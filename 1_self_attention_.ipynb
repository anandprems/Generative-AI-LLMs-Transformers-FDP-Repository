```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)
```

```python
Q = np.random.rand(3, 4)
K = np.random.rand(3, 4)
V = np.random.rand(3, 4)

print(Q)
print(K)
print(V)
```

```python
scores = np.dot(Q, K.T)
print(scores)
```

```python
attention_weights = softmax(scores)
print(attention_weights)
```

```python
output = np.dot(attention_weights, V)
print(output)
```
