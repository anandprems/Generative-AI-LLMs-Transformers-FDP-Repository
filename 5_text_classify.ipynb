```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
text = "This FDP session is extremely useful and well explained."

result = classifier(text)

print("Text:", text)
print("Prediction:", result)
```

```python
texts = [
    "I love teaching students",
    "The lecture was boring and confusing",
    "This topic is difficult but interesting",
    "I hate exams"
]

results = classifier(texts)

for text, res in zip(texts, results):
    print(f"\"{text}\" → {res['label']} (confidence: {res['score']:.2f})")
```
