```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
# positive context
prompt_1 = "The student solved the problem because"

output_1 = generator(
    prompt_1,
    max_length=40,
    temperature=0.7,
    top_k=50
)

print("Prompt:", prompt_1)
print("Output:", output_1[0]["generated_text"])
```

```python
# negative context
prompt_2 = "The student failed the problem because"

output_2 = generator(
    prompt_2,
    max_length=40,
    temperature=0.7,
    top_k=50
)

print("\nPrompt:", prompt_2)
print("Output:", output_2[0]["generated_text"])
```
